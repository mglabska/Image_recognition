```python
# Run some predictions on pictures from the internet

#Imports
import matplotlib.pyplot as plt
import numpy as np
from keras.models import load_model
from keras.preprocessing.image import ImageDataGenerator
```

```python
# Load model
new_model = load_model('models\\my_model.h5_vgg')
```

```python
# Generate data
image1 = ImageDataGenerator(rescale=1./255).flow_from_directory('C:\\Users\\admin\\SDA\\ComputerVision\\Projekt\\microsoft-catsvsdogs-dataset\\predictions\\samples', target_size=[100,100])
```

```python
#Check class indices
image1.class_indices
# In my model 0 indicates cats, 1 - dogs.
```

```python
new_model.predict(image1)
```

```python
predict_1 = new_model.predict(image1)
predictions = np.round(predict_1).astype(int)
print(predictions)
# Seems that the first image represents a dog, the second one - a cat.
```

```python
# In a loop
models = ['models\\my_model.h5_vgg', 'models\\my_model.h5_splitted']
```
